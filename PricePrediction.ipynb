```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
train = pd.read_csv("/Users/grayfloyd/Desktop/projects/PricePrediction/train.csv")
test = pd.read_csv("/Users/grayfloyd/Desktop/projects/PricePrediction/test.csv")
train
```

```python
pd.set_option('display.max_columns', None)
train
```

```python
columns_with_missing_values = train.columns[train.isnull().any()]
missing_values_count = train[columns_with_missing_values].isnull().sum()
missing_values_count
```

```python
train = train.drop(["Alley", "MasVnrType", "FireplaceQu", "PoolQC", "Fence", "MiscFeature", "Id"], axis=1)
train
```

```python
numerical_columns = train.select_dtypes(include='number').columns
train[numerical_columns] = train[numerical_columns].fillna(train[numerical_columns].mean())
```

```python
categorical_columns = train.select_dtypes(include='object').columns
train[categorical_columns] = train[categorical_columns].fillna(train[categorical_columns].mode().iloc[0])
```

```python
import matplotlib.pyplot as plt

numerical_columns1 = train.select_dtypes(include='number')
correlation_matrix = numerical_columns1.corr()
plt.figure(figsize=(30, 28))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

for column in train.select_dtypes(include='object'):
    model = ols('SalePrice ~ ' + column, data=train).fit()
    anova_table = sm.stats.anova_lm(model)
    print(f"ANOVA for {column}:")
    print(anova_table)
    print()
```
